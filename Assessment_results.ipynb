###### consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv


###### if  10000<codekata score<15000   (Reached_expectations.csv)



###### if  7000<codekata score<10000   (Needs_Improvement.csv)


###### if  codekate score < 7000        (Unsatisfactory.csv)

```python
#import pandas module
import pandas as pd
#read the csv file into a dataframe
df1 = pd.read_csv("college_1.csv")
df_e=df1[df1['CodeKata Score']>15000]
df_e.to_csv('exceeded.csv')
df_r=df1[(df1['CodeKata Score'] >= 10000) & (df1['CodeKata Score'] <= 15000)]
df_r.to_csv('reached.csv')
df_n=df1[(df1['CodeKata Score'] >= 7000) & (df1['CodeKata Score'] <= 10000)]
df_n.to_csv('needs_improve')
df_u=df1[df1['CodeKata Score']<7000]
df_u.to_csv('unsatis.csv')
```

```python
#import pandas module
import pandas as pd
#read the csv file into a dataframe
df2 = pd.read_csv("college_2.csv")
df_e1=df2[df2['CodeKata Score']>15000]
df_e1.to_csv('exceeded1.csv')
df_r1=df2[(df2['CodeKata Score'] >= 10000) & (df1['CodeKata Score'] <= 15000)]
df_r1.to_csv('reached1.csv')
df_n1=df2[(df2['CodeKata Score'] >= 7000) & (df1['CodeKata Score'] <= 10000)]
df_n1.to_csv('needs_improve1.csv')
df_u1=df2[df2['CodeKata Score']<7000]
df_u1.to_csv('unsatis1.csv')
```
